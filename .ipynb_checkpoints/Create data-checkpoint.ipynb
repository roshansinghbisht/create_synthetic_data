```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('paper_texture.jpg')
plt.imshow(image)
```

```python
def draw_character_bounding_box(image, character, org, font, fontScale, color, thickness):
    image = cv2.putText(image, character, org, font, 
                       fontScale, color, thickness, cv2.LINE_AA)
    
    char_size = cv2.getTextSize(character, font, fontScale, thickness)
#     image = cv2.rectangle(image, (org[0], org[1]-char_size[0][1]), (org[0]+char_size[0][1], org[1]), (0,255,0),5)
    image = cv2.rectangle(image, (org[0], org[1]-char_size[0][1]), (org[0]+char_size[0][0], org[1]), (0,255,0),5)
    
    return image, char_size
```

```python
# font
font = cv2.FONT_HERSHEY_SIMPLEX
  
# org
org = (400, 800)
  
# fontScale
fontScale = 5
   
# Blue color in BGR
color = (255, 0, 0)
  
# Line thickness of 2 px
thickness = 20
```

```python
chars = ['U', '1', '2']#, '3', '4', '5', '6', '7', '8', '9', '0']
for char in chars:
    image, char_size = draw_character_bounding_box(image, char, org, font, fontScale, color, thickness)
    org = (org[0]+char_size[0][0], org[1])
#     org = (org[0]+char_size[0][1], org[1])

# plt.imshow(image)
# #Display the image
cv2.imshow('All contours with bounding box', image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Save image
cv2.imwrite("out.jpg", image)
```
