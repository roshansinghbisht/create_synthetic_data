```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('paper_texture.jpg')
plt.imshow(image)
```

```python
# font
font = cv2.FONT_HERSHEY_SIMPLEX
  
# org
org = (500, 400)
  
# fontScale
fontScale = 10
   
# Blue color in BGR
color = (255, 0, 0)
  
# Line thickness of 2 px
thickness = 20
```

```python
image = cv2.putText(image, 'OpenCV', org, font, 
                    fontScale, color, thickness, cv2.LINE_AA)
    
char_size = cv2.getTextSize('OpenCV',cv2.FONT_HERSHEY_SIMPLEX, 10,20)
char_size
```

```python
image = cv2.rectangle(image, (org[0], org[1]-char_size[0][1]), (org[0]+char_size[0][0], org[1]), (0,255,0),5)                                                                                             

# Displaying the image
plt.imshow(image)
```
